# Question 5 (I)

```python
# consider Alternate Hypothesis,p < 5% which means that Mozilla has a less than 5% share of the market
# otherwise Null Hypothesis,p>=5%  which means that Mozilla has more than or equal to 5% share of the market
# first calculation for findng the Z score
Z_score = (0.046-0.05)/(np.sqrt((0.05*(1-0.05))/2000))
print('Z_score is:',Z_score)
```

```python
# calculation for finding the probability
P = 1-stats.norm.cdf(Z_score)
P
```

it is greater than 0.05, therefore we can conclude that our null hypothesis is correct, means that Mozilla has more than or equal to 5% share of the market
